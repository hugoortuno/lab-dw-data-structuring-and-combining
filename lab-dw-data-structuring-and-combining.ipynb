```python
## 3. Combine the Datasets

# Concatenate the datasets if they have the same structure
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
# 4. Clean the Data

# Ensure 'text_column' is an actual column in your DataFrame
if 'text_column' in combined_df.columns:
    combined_df['text_column'] = combined_df['text_column'].str.lower()

# Optional: Check the cleaned data
print(combined_df.head())
print(combined_df.info())
```

```python
# 5. Verify Data Quality

# Check for remaining missing values
print(combined_df.isnull().sum())

# Check data types
print(combined_df.dtypes)
```

```python
# 6. Save the Cleaned Data

# Save the cleaned dataset
cleaned_file_path = 'cleaned_combined_data.csv'
combined_df.to_csv(cleaned_file_path, index=False)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

import pandas as pd

# Cargar el dataset
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df = pd.read_csv(url)

# Inspeccionar los datos
print("Primeras filas del dataset:")
print(df.head())
print("\nTipos de datos y valores faltantes:")
print(df.info())
print("\nNombres de las columnas en el DataFrame:")
print(df.columns)

# Eliminar duplicados
df = df.drop_duplicates()

# Manejar valores faltantes
# Reemplazar valores faltantes en columnas categóricas con 'Unknown'
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna('Unknown')

# Reemplazar valores faltantes en columnas numéricas con la mediana
for col in df.select_dtypes(include=['number']).columns:
    df[col] = df[col].fillna(df[col].median())

# Convertir tipos de datos
# Convertir columnas de fechas a datetime si es aplicable
date_columns = ['date_column']  # Reemplazar con columnas de fecha reales
for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Convertir columnas categóricas a tipo categoría
categorical_columns = ['categorical_column1', 'categorical_column2']  # Reemplazar con columnas categóricas reales
for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].astype('category')

# Normalizar columnas de texto
text_columns = ['text_column1', 'text_column2']  # Reemplazar con columnas de texto reales
for col in text_columns:
    if col in df.columns:
        df[col] = df[col].str.lower()

# Guardar datos limpios en un nuevo archivo CSV
df.to_csv('cleaned_marketing_customer_analysis.csv', index=False)

# Realizar un análisis exploratorio de datos (EDA)
print("\nEDA - Conteo de valores para columnas categóricas:")
for col in categorical_columns:
    if col in df.columns:
        print(f"\nConteo de valores para {col}:")
        print(df[col].value_counts())

# Matriz de correlación solo con columnas numéricas
print("\nMatriz de correlación:")
numeric_df = df.select_dtypes(include=['number'])  # Seleccionar solo columnas numéricas
print(numeric_df.corr())

# Visualizaciones (opcional)
import matplotlib.pyplot as plt
import seaborn as sns

# Ejemplo: Distribución de una columna numérica
numeric_column = 'numeric_column'  # Reemplazar con una columna numérica real
if numeric_column in df.columns:
    sns.histplot(df[numeric_column])
    plt.title('Distribución de la Columna Numérica')
    plt.show()

# Ejemplo: Gráfico de conteo para una columna categórica
categorical_column = 'categorical_column1'  # Reemplazar con una columna categórica real
if categorical_column in df.columns:
    sns.countplot(x=categorical_column, data=df)
    plt.title('Conteo de la Columna Categórica')
    plt.show()

```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Paso 1: Cargar y Examinar los Datos

import pandas as pd

# Cargar el dataset
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df = pd.read_csv(url)

# Mostrar los nombres de las columnas y las primeras filas
print("Nombres de las columnas:", df.columns)
print("Primeras filas del DataFrame:")
print(df.head())
```

```python
# Paso 2: Crear la Tabla Dinámica

# Limpiar nombres de columnas
df.columns = df.columns.str.strip()

# Asegúrate de que los nombres de las columnas sean correctos
# Supongamos que las columnas relevantes son 'Customer_Lifetime_Value', 'Gender', y 'Education_Level'
pivot_table = df.pivot_table(
    values='customer_lifetime_value',  # Nombre de la columna para el CLV
    index='gender',  # Nombre de la columna para el género
    columns='education',  # Nombre de la columna para el nivel educativo
    aggfunc='mean'
)

# Redondear el promedio del CLV a 2 decimales
pivot_table = pivot_table.round(2)

# Mostrar la tabla dinámica
print("Tabla dinámica mostrando el promedio de CLV por género y nivel educativo:")
print(pivot_table)
```

Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Your code goes here

import pandas as pd

# Crear un DataFrame en formato largo
df_long = pd.DataFrame({
    'ID': [1, 1, 1, 1, 2, 2, 2, 2],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02'],
    'Variable': ['Revenue', 'Revenue', 'Expense', 'Expense', 'Revenue', 'Revenue', 'Expense', 'Expense'],
    'Value': [1000, 1100, 500, 550, 2000, 2200, 1000, 1100]
})

# Convertir a formato ancho
df_wide = df_long.pivot_table(index=['ID', 'Date'], columns='Variable', values='Value').reset_index()

print(df_wide)
```
